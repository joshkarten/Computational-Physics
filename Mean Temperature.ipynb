```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
datT = np.loadtxt('stockholm_daily_mean_temperature.csv',delimiter=',',skiprows=1,usecols=[3])
datd = np.loadtxt('stockholm_daily_mean_temperature.csv',delimiter=',',skiprows=1,usecols=[0],dtype=str)
datd_ =np.array([[int(x) for x in dt.split('-')] for dt in datd])
datd = datd_
```

```python
lastYear = datd[-1][0] 
firstYear = datd[0][0]
```

```python
x=np.linspace(firstYear,lastYear,datT.size)
fig,ax = plt.subplots()
ax.scatter(x,datT, marker='+')
ax.set_xlabel('Year')
ax.set_ylabel(r'Temperaturn ($\degree$ C)')
ax.set_title(r'Daily temperature for Stockholm in $\degree$ C ')
plt.show()
```

```python
i=np.where(datd==firstYear)
numPos = np.zeros(lastYear-firstYear+1)
posNorm = np.zeros(lastYear-firstYear+1)
newPos = np.zeros(lastYear-firstYear)
for year in range(firstYear,lastYear+1):
    for i in np.where(datd[:,0]==year)[0]:
        if datT[i]>0:
            numPos[year-firstYear]+=1
    posNorm[year-firstYear]= numPos[year-firstYear]/np.size(np.where(datd[:,0]==year)[0])
for i in range(lastYear-firstYear):
    newPos[i] =( numPos[i]/np.size(np.where(datd[:,0]==(i+firstYear))[0])+numPos[i+1]/np.size(np.where(datd[:,0]==(i+firstYear+1))[0]))/2
fig,ax = plt.subplots()
ax.plot(np.linspace(firstYear, lastYear, lastYear-firstYear),numPos[:-1])
ax.set_ylabel("Number of Positive Days")
ax.set_xlabel("Year")
ax.set_title(r"Number of Days with Temperature Greater than 0$\degree$ C over Time in Stockholm")
fig,ax=plt.subplots()
ax.plot(np.linspace(firstYear, lastYear, lastYear-firstYear),posNorm[:-1])
ax.set_ylabel("Number of Positive Days over Number of Days")
ax.set_xlabel("Year")
ax.set_title(r"Number of Days with Temperature Greater than 0$\degree$ C over Time in Stockholm")

fig,ax=plt.subplots()
years = list(range(firstYear,2024,1))
ax.plot(np.linspace(firstYear, lastYear-2, lastYear-firstYear-1),newPos[:-1])
ax.set_ylabel("2 Year Rolling Average of Number of Positive Days")
ax.set_title(r"Percentage of Days with Temperature Greater than $0\degree$C over Time in Stockholm")

ax.set_xlabel("Year")
plt.show()
```

By looking at the above graphical representation of the daily mean temperature of Stockholm, we clearly see the upwards trends of the mean temperature.
